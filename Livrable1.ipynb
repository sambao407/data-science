```python
from skimage import io
import matplotlib.pyplot as plt

image = io.imread('pioupiou.png')

plt.hist(image.ravel(), bins = 256, color = 'orange', )
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.legend(['Total'])
plt.show()
```
